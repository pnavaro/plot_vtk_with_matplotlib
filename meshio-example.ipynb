## Read the file 

```python
import numpy as np
import meshio
```

```python
reader = meshio.read("pep1_7_OmegaV.vtu")
```

## Get mesh data

### Points

```python
reader.points
```

### Cells

```python
reader.cells
```

```python
x = reader.points
```

```python
triangles =  reader.cells['triangle']
```

### Field values

```python
reader.point_data
```

```python
u = reader.point_data['abs_term_u']
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10.0, 6.0)
```

```python
# Mesh
plt.figure(figsize=(8, 8))
plt.triplot(x[:,0], x[:,1], triangles)
plt.gca().set_aspect('equal')
```

```python
# abs_u 
plt.figure(figsize=(8, 8))
plt.tricontour(x[:,0], x[:,1], triangles, u, 16);
```

```python
plt.figure(figsize=(8, 8))
plt.tricontourf(x[:,0], x[:,1], triangles, u, 16);
```
